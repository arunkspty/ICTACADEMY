# 4. On average which region has the highest
a. GDP per capita (current US$)
b. International trade: Imports (million US$)

```python
df1=data.groupby('Region')['GDP per capita (current US$)']
df2=df1.describe()
df2['mean'].mean().round(2)
```

```python
df2
```

```python
plt.figure(figsize=(8,6))
data.groupby('Region')['GDP per capita (current US$)'].mean().plot.bar(color='green')
plt.title("Average Regional GDP")
plt.ylabel("Average GDP")
plt.show()
print('On the basis of the average GDP w.r.t the regions, $16109.07 is the average GDP of all the 22 regions. North America has the   highest average GDP.')
```

```python
df3=data.groupby('Region')['International trade: Imports (million US$)']
df4=df3.describe()
df4['mean'].mean().round(2)
```

```python
df4
```

```python
plt.figure(figsize=(8,6))
data.groupby('Region')['International trade: Imports (million US$)'].mean().plot.bar(color='red')
plt.title("Average Regionl International Import Trade")
plt.ylabel("Average International Import trade")
plt.show()
print('On the basis of the average international imports trade w.r.t the regions, $104060.84 is the average imports trade of all the  22 regions. North America has the highest international import trade.')
```
