```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

 # Reading the file and Check for the null values present in the dataset

```python
data=pd.read_csv('cars_data.csv')
data=data.drop(data[data["Make"]=="?"].index)
data
```

# Plot a bar graph of male vs female buyers participated in the sales

```python
y=data["Buyer Gender"].value_counts()
A=[y["Male"],y["Female"]]
```

```python
plt.figure(figsize=(8,6))
ax = sns.barplot(x=["Male","Female"], y=A)
plt.grid()
plt.title('Male Vs Female buyers participated in the Sales',fontsize=15,fontweight='bold')
plt.xlabel('Buyer Gender',fontsize=12,fontweight='bold')
plt.ylabel('Count',fontsize=12,fontweight='bold')
plt.show()
```
